```python
#Creating Movies RDD and then storing the header in one variable:
movies=sc.textFile('/FileStore/tables/movies.csv')
headerVal=movies.first()
headerVal
```

```python
#Removing the header from the data and storing the total value in another RDD
moviesData=movies.filter(lambda x: x != headerVal)
moviesData.take(10)
```

```python
#Creating ratings RDD:
ratings=sc.textFile('/FileStore/tables/ratings.csv')
ratings.take(4)
```

```python
#Storing the ratings header in another variable:
ratingsHeader=ratings.first()
ratingsHeader
```

```python
#Removing the header and storing the rest values in an RDD:
ratingsData=ratings.filter(lambda x: x!=ratingsHeader)
ratingsData.take(4)
```

```python
#Storing the movie_id and movie_name together in an RDD:
movieDataSplit=moviesData.map(lambda x:(x.split(',')[0],x.split(',')[1]))
movieDataSplit.collect()
```

```python
#Creating a dictionary so that it can be passed to the broadcast variable:
movieDataSplitDict=dict()
for i in movieDataSplit.collect():
  movieDataSplitDict[i[0]]=i[1]
movieDataSplitDict
```

```python
#Creating the broadcast variable to store the movies dictionary created previously:
movie_broadCast = sc.broadcast(movieDataSplitDict)
movie_broadCast.value
```

```python
#Creating an RDD composed of movie_id and their counts, as found in the ratings RDD:
ratingsCount=ratingsData.map(lambda x:(x.split(',')[1],1)).reduceByKey(lambda a,b:a+b)
ratingsCount.take(10)
```

```python
#checking which column of ratingsCount RDD giving the count field:
ratingsCount.map(lambda x:x[1]).take(2) 
```

```python
ratingsCount.map(lambda x: movie_broadCast.value).take(2)
```

```python
#checking after mapping the ratingsCount RDD with broadcast variable to get the movie_name:
ratingsCount.map(lambda x: movie_broadCast.value[x[0]]).take(2)
```

```python
#Creating an RDD having movie_name and its count,by using the broadcast variable:
movie_count=ratingsCount.map(lambda x: (movie_broadCast.value[x[0]],x[1]))
movie_count.take(4)
```

```python
#Print the top 100 movies where the count is more or the data has been arranged in descending order:
movie_count.sortBy(lambda x: x[1], ascending= False).take(100)
```
